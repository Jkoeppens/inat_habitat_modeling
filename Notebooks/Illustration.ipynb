```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import xgboost as xgb
from sklearn.metrics import (
    roc_curve, auc,
    precision_recall_curve, average_precision_score,
    confusion_matrix, classification_report
)

from sklearn.calibration import calibration_curve

# -----------------------------------------------------------
# 1) Pfade
# -----------------------------------------------------------

CSV_PATH = "/Volumes/Data/iNaturalist/features/Macrolepiota_procera/inat_with_climatology_Macrolepiota_procera_vs_Parus_major.csv"
MODEL_PATH = "/Volumes/Data/iNaturalist/outputs/macrolepiota_procera/model_Macrolepiota_procera_vs_Parus_major.json"
OUTPUT_DIR = "/Volumes/Data/iNaturalist/outputs/macrolepiota_procera/plots"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# -----------------------------------------------------------
# 2) CSV laden
# -----------------------------------------------------------

df = pd.read_csv(CSV_PATH)

# Entferne nicht-numerische Spalten, die das Modell nicht hatte
drop_cols = ["species", "taxon_id", "date", "latitude", "longitude"]
df = df.drop(columns=drop_cols)

# y & X definieren
y = df["label"]
X = df.drop(columns=["label"])

# Train/test split (reproduzierbar)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# -----------------------------------------------------------
# 3) XGBoost laden
# -----------------------------------------------------------

model = xgb.XGBClassifier()
model.load_model(MODEL_PATH)

# Wahrscheinlichkeiten berechnen
y_proba = model.predict_proba(X_test)[:, 1]

# -----------------------------------------------------------
# 4) Plot-Funktionen (Custom Lift + Custom Cumulative Gain)
# -----------------------------------------------------------

def plot_lift_curve(y_true, y_proba, save_path, bins=10):
    df = pd.DataFrame({"y": y_true, "p": y_proba})
    df["bin"] = pd.qcut(df["p"], q=bins, duplicates="drop")

    grouped = df.groupby("bin")
    pos_rate = grouped["y"].mean()

    overall_rate = df["y"].mean()
    lift = pos_rate / overall_rate

    plt.figure(figsize=(7,5))
    plt.plot(np.arange(1, len(lift)+1), lift.values, marker="o")
    plt.axhline(1, color="k", linestyle="--")
    plt.title("Lift Curve")
    plt.xlabel("Probability Decile (High â†’ Low)")
    plt.ylabel("Lift")
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.savefig(save_path, dpi=300)
    plt.close()


def plot_cumulative_gain(y_true, y_proba, save_path):
    df = pd.DataFrame({"y": y_true, "p": y_proba})
    df = df.sort_values("p", ascending=False)

    df["cum_pos"] = (df["y"] == 1).cumsum()
    df["gain"] = df["cum_pos"] / df["y"].sum()
    df["cases_pct"] = np.arange(1, len(df)+1) / len(df)

    plt.figure(figsize=(7,5))
    plt.plot(df["cases_pct"], df["gain"], lw=2)
    plt.plot([0,1], [0,1], "k--", label="Baseline")
    plt.xlabel("Proportion of Samples")
    plt.ylabel("Cumulative Gain")
    plt.title("Cumulative Gains Chart")
    plt.grid(alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.savefig(save_path, dpi=300)
    plt.close()

# -----------------------------------------------------------
# 5) ROC Curve
# -----------------------------------------------------------

fpr, tpr, th = roc_curve(y_test, y_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, lw=2, label=f"AUC = {roc_auc:.3f}")
plt.plot([0,1], [0,1], "k--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "roc_curve.png"), dpi=300)
plt.close()

# -----------------------------------------------------------
# 6) Precisionâ€“Recall Curve
# -----------------------------------------------------------

prec, rec, th = precision_recall_curve(y_test, y_proba)
ap = average_precision_score(y_test, y_proba)

plt.figure(figsize=(6,6))
plt.plot(rec, prec, lw=2, label=f"AP = {ap:.3f}")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precisionâ€“Recall Curve")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "precision_recall.png"), dpi=300)
plt.close()

# -----------------------------------------------------------
# 7) Confusion Matrix (default threshold 0.5)
# -----------------------------------------------------------

y_pred = (y_proba >= 0.5).astype(int)
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix (threshold = 0.5)")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "confusion_matrix.png"), dpi=300)
plt.close()

# -----------------------------------------------------------
# 8) Histogram der Scores
# -----------------------------------------------------------

plt.figure(figsize=(7,5))
sns.histplot(y_proba, bins=30, kde=True)
plt.title("Histogram of Predicted Probabilities")
plt.xlabel("Score")
plt.ylabel("Count")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "score_histogram.png"), dpi=300)
plt.close()

# -----------------------------------------------------------
# 9) Calibration Curve
# -----------------------------------------------------------

prob_true, prob_pred = calibration_curve(y_test, y_proba, n_bins=10)

plt.figure(figsize=(6,6))
plt.plot(prob_pred, prob_true, marker="o")
plt.plot([0,1], [0,1], "k--")
plt.xlabel("Predicted Probability")
plt.ylabel("True Probability")
plt.title("Calibration Curve")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "calibration_curve.png"), dpi=300)
plt.close()

# -----------------------------------------------------------
# 10) Lift & Cumulative Gain
# -----------------------------------------------------------

plot_lift_curve(y_test, y_proba, os.path.join(OUTPUT_DIR, "lift_curve.png"))
plot_cumulative_gain(y_test, y_proba, os.path.join(OUTPUT_DIR, "cumulative_gains.png"))

print("\nðŸŽ‰ Alle Modell-Diagramme erfolgreich gespeichert in:")
print(OUTPUT_DIR)
```

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
import joypy
import pandas as pd
import glob
import os

# ---------------------------------------------------
# 1) Pfade zu deinen Jahres-Suitability-TIFFs
# ---------------------------------------------------
folder = "/Volumes/Data/iNaturalist/outputs/macrolepiota_procera"

# Dateinamen wie:
#   suitability_2021_MONTHLY_Macrolepiota_procera_vs_Parus_major.tif
tiff_paths = sorted(glob.glob(os.path.join(folder, "suitability_20*_MONTHLY_*.tif")))

print("Gefundene Dateien:")
for p in tiff_paths:
    print(" â†’", p)

# ---------------------------------------------------
# 2) Raster laden und Pixelwerte extrahieren
# ---------------------------------------------------
def load_values(path):
    with rasterio.open(path) as src:
        arr = src.read(1).astype(float)
    arr = arr[np.isfinite(arr)]          # NaN entfernen
    arr = arr[(arr >= 0) & (arr <= 1)]   # Wertebereich filtern
    return arr

data = {}
for path in tiff_paths:
    year = int(os.path.basename(path).split("_")[1])  # z.B. '2021'
    data[year] = load_values(path)

# ---------------------------------------------------
# 3) KDEs einzeln plotten (Kontrolle)
# ---------------------------------------------------
plt.figure(figsize=(10,5))

for year, vals in data.items():
    kde = gaussian_kde(vals)
    xs = np.linspace(0, 1, 300)
    plt.plot(xs, kde(xs), label=str(year))

plt.title("KDE pro Jahr â€“ Suitability-Verteilungen")
plt.xlabel("Suitability")
plt.ylabel("Dichte")
plt.legend()
plt.show()

# ---------------------------------------------------
# 4) Ridgeline-Daten in ein DataFrame packen
# ---------------------------------------------------
records = []
for year, vals in data.items():
    df = pd.DataFrame({
        "suitability": vals,
        "year": str(year)
    })
    records.append(df)

df_all = pd.concat(records, ignore_index=True)

# ---------------------------------------------------
# 5) Ridgeline-Plot mit joypy
# ---------------------------------------------------
plt.figure(figsize=(10,8))
joypy.joyplot(
    df_all,
    by="year",
    column="suitability",
    colormap=plt.cm.plasma,       # spÃ¤ter Gradient anpassbar
    bins=40,
    fade=True,
    figsize=(10,8),
    range_style="own",
    kind="kde"
)

plt.title("Ridgeline: Suitability-Verteilung 2020â€“2024", fontsize=16)
plt.xlabel("Suitability")
plt.tight_layout()
plt.show()
```

```python
#Ridgeline Grafen machen aus Jahres-Suitability Tiffs

import rasterio
import numpy as np
from scipy.stats import gaussian_kde
import plotly.graph_objects as go
import os

# ---------------------------------------------------
# 1) Pfade zu deinen Jahres-Suitability-TIFFs
# ---------------------------------------------------

folder = "/Volumes/Data/iNaturalist/outputs/macrolepiota_procera"
years = [2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]

# Dateiname-Muster anpassen:
def path_for_year(year):
    return f"{folder}/suitability_{year}_MONTHLY_Macrolepiota_procera_vs_Parus_major.tif"


# ---------------------------------------------------
# 2) Funktion: Raster einlesen & sampeln
# ---------------------------------------------------

def load_and_sample(path, max_pixels=120000):
    with rasterio.open(path) as src:
        arr = src.read(1).astype("float32")
    arr = arr[np.isfinite(arr)]

    if len(arr) > max_pixels:
        idx = np.random.choice(len(arr), max_pixels, replace=False)
        arr = arr[idx]

    return arr


# ---------------------------------------------------
# 3) KDE fÃ¼r ein Jahr berechnen
# ---------------------------------------------------

def compute_kde(values, bins=300):
    kde = gaussian_kde(values)
    xs = np.linspace(0, 1, bins)
    ys = kde(xs)
    return xs, ys / ys.max()   # Normalisiert (optisch schÃ¶ner)


# ---------------------------------------------------
# 4) 3D Ridgeline in Plotly
# ---------------------------------------------------

fig = go.Figure()

for i, year in enumerate(years):
    tif_path = path_for_year(year)
    if not os.path.exists(tif_path):
        print("âš ï¸ Datei fehlt:", tif_path)
        continue

    vals = load_and_sample(tif_path, max_pixels=100000)

    xs, zs = compute_kde(vals)

    ys = np.full_like(xs, i)   # Y-Achse = Jahresindex

    fig.add_trace(go.Scatter3d(
        x=xs,
        y=ys,
        z=zs,
        mode='lines',
        line=dict(
            width=6,
            color=i,
            colorscale="Plasma"   # lilaâ†’gelb wie im Beispiel
        ),
        name=str(year)
    ))

# Layout
fig.update_layout(
    title="3D Ridgeline der Habitat-Suitability (KDE pro Jahr)",
    scene=dict(
        xaxis_title="Suitability",
        yaxis_title="Jahr",
        zaxis_title="Dichte (KDE)",
        yaxis=dict(
            tickmode="array",
            tickvals=list(range(len(years))),
            ticktext=[str(y) for y in years]
        )
    ),
    width=1000,
    height=700
)

fig.show()
```

```python
#letzte funktionierende histogramm-animation, leider loopt das ergebenis einmal

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
import xgboost as xgb
from sklearn.model_selection import train_test_split
from PIL import Image, ImageSequence

# -----------------------------------------------------------
# 1) Daten laden
# -----------------------------------------------------------
csv_path = "/Volumes/Data/iNaturalist/features/Macrolepiota_procera/inat_with_climatology_Macrolepiota_procera_vs_Parus_major.csv"
df = pd.read_csv(csv_path)

feature_cols = [c for c in df.columns if c.startswith("m") and "coverage" not in c]

X = df[feature_cols]
y = df["label"].astype(int)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# -----------------------------------------------------------
# 2) Modell laden
# -----------------------------------------------------------
model = xgb.XGBClassifier()
model.load_model("/Volumes/Data/iNaturalist/outputs/macrolepiota_procera/model_Macrolepiota_procera_vs_Parus_major.json")

y_prob = model.predict_proba(X_test)[:, 1]

# Klassen trennen
pos = y_prob[y_test == 1]
neg = y_prob[y_test == 0]

# Histogramm
bins = np.linspace(0, 1, 25)
pos_hist, _ = np.histogram(pos, bins=bins)
neg_hist, _ = np.histogram(neg, bins=bins)

max_height = max(pos_hist.max(), neg_hist.max())
bin_centers = 0.5 * (bins[:-1] + bins[1:])

# -----------------------------------------------------------
# 3) Plot Setup
# -----------------------------------------------------------
fig, ax = plt.subplots(figsize=(7, 4))
ax.set_title("Probability Histogram â€“ Growing Animation")
ax.set_xlabel("Predicted Probability")
ax.set_ylabel("Count")
ax.set_xlim(0, 1)
ax.set_ylim(0, max_height * 1.1)

pos_bars = ax.bar(bin_centers - 0.01, np.zeros_like(pos_hist),
                  width=0.02, color="#4477AA", label="Pilz")
neg_bars = ax.bar(bin_centers + 0.01, np.zeros_like(neg_hist),
                  width=0.02, color="#EE7733", label="Meise")
ax.legend()

# -----------------------------------------------------------
# 4) Animation
# -----------------------------------------------------------
frames = 40

def update(frame):
    scale = frame / (frames - 1)
    for r, h in zip(pos_bars, pos_hist): r.set_height(h * scale)
    for r, h in zip(neg_bars, neg_hist): r.set_height(h * scale)
    return list(pos_bars) + list(neg_bars)

anim = FuncAnimation(fig, update, frames=frames, interval=50, blit=True)
temp_gif = "prob_temp.gif"
anim.save(temp_gif, writer=PillowWriter(fps=30))
plt.close(fig)

# -----------------------------------------------------------
# 5) Loop entfernen â†’ Finale GIF
# -----------------------------------------------------------
def remove_loop(in_path, out_path):
    img = Image.open(in_path)
    frames = []
    durations = []
    for frame in ImageSequence.Iterator(img):
        frames.append(frame.copy())
        durations.append(frame.info.get("duration", 40))
    frames[0].save(out_path, save_all=True,
                   append_images=frames[1:],
                   duration=durations,
                   loop=1)

remove_loop(temp_gif, "probability_histogram_once.gif")
print("FERTIG â†’ probability_histogram_once.gif")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
import xgboost as xgb

# -----------------------------------------------------------
# Daten laden
# -----------------------------------------------------------
csv_path = "/Volumes/Data/iNaturalist/features/Macrolepiota_procera/inat_with_climatology_Macrolepiota_procera_vs_Parus_major.csv"
df = pd.read_csv(csv_path)

feature_cols = [c for c in df.columns if c.startswith("m") and "coverage" not in c]

# Modell laden
model = xgb.XGBClassifier()
model.load_model("/Volumes/Data/iNaturalist/outputs/macrolepiota_procera/model_Macrolepiota_procera_vs_Parus_major.json")

# Feature Importances
importance = model.feature_importances_
idx = np.argsort(importance)[::-1][:10]

top_features = [feature_cols[i] for i in idx]
top_importances = importance[idx]

# -----------------------------------------------------------
# Plot Setup â€” schmal & hoch
# -----------------------------------------------------------
fig, ax = plt.subplots(figsize=(6, 8))   # hÃ¶her, schmaler
plt.title("Top 10 Feature Importances (Animated)", fontsize=16)

y_pos = np.arange(len(top_features))

bars = ax.barh(y_pos, np.zeros_like(top_importances), color="#4477AA")

ax.set_yticks(y_pos)
ax.set_yticklabels(top_features, fontsize=10)
ax.invert_yaxis()  # wichtig: wichtigstes Feature oben
ax.set_xlim(0, top_importances.max() * 1.15)

# -----------------------------------------------------------
# Animation
# -----------------------------------------------------------
FRAMES = 40
FPS = 30
PAUSE_SECONDS = 30
PAUSE_FRAMES = PAUSE_SECONDS * FPS
TOTAL_FRAMES = FRAMES + PAUSE_FRAMES

def update(frame):
    # wÃ¤hrend der Pause â†’ alles voll aufgebaut
    if frame >= FRAMES:
        scale = 1.0
    else:
        scale = frame / (FRAMES - 1)

    for rect, h in zip(bars, top_importances):
        rect.set_width(h * scale)

    return bars

anim = FuncAnimation(fig, update, frames=TOTAL_FRAMES, interval=1000/FPS, blit=True)

# -----------------------------------------------------------
# GIF speichern
# -----------------------------------------------------------
writer = PillowWriter(fps=FPS)
anim.save("feature_importances_pause.gif", writer=writer)

plt.close(fig)
print("Fertig! â†’ feature_importances_pause.gif")
```

```python
#feature importance graph mit labels (fingiert)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
import xgboost as xgb

# ------------------------------------------
# 1) Modell & Daten laden
# ------------------------------------------
CSV = "/Volumes/Data/iNaturalist/features/Macrolepiota_procera/inat_with_climatology_Macrolepiota_procera_vs_Parus_major.csv"
MODEL = "/Volumes/Data/iNaturalist/outputs/macrolepiota_procera/model_Macrolepiota_procera_vs_Parus_major.json"

df = pd.read_csv(CSV)

# Feature-Spalten wie im Training
feature_cols = [c for c in df.columns if c.startswith("m") and "coverage" not in c]

model = xgb.XGBClassifier()
model.load_model(MODEL)

importance = model.feature_importances_
importance_df = pd.DataFrame({
    "feature": feature_cols,
    "importance": importance
})

# Top 10 Features
imp10 = importance_df.sort_values("importance", ascending=False).head(10)

# ------------------------------------------
# 2) Manueller Label-Mapping (editierbar!)
# ------------------------------------------
label_map = {
    "m10_ndvi_mean": "NDVI (Okt)",
    "m10_ndwi_mean": "NDWI (Okt)",
    "m10_moran_ndvi": "NDVI Moran (Okt)",
    "m10_geary_ndvi": "NDVI Geary (Okt)",
    "m10_moran_ndwi": "NDWI Moran (Okt)",
    "m10_geary_ndwi": "NDWI Geary (Okt)",
    "m09_ndvi_mean": "NDVI (Sep)",
    "m09_ndwi_mean": "NDWI (Sep)",
    "m11_ndvi_mean": "NDVI (Nov)",
    "m11_ndwi_mean": "NDWI (Nov)",
}

# Wenn Feature nicht im Mapping â†’ Originalname
imp10["label"] = imp10["feature"].map(lambda x: label_map.get(x, x))

# ------------------------------------------
# 3) Plot Setup
# ------------------------------------------
fig, ax = plt.subplots(figsize=(5, 8))   # HOCHFORMAT
plt.title("Top 10 Feature Importances â€“ Animation")

y_pos = np.arange(len(imp10))
heights = imp10["importance"].values

bars = ax.barh(y_pos, np.zeros_like(heights), color="#4477AA")
ax.set_yticks(y_pos)
ax.set_yticklabels(imp10["label"], fontsize=10)

ax.invert_yaxis()  # wichtig: wichtigstes Feature oben
ax.set_xlim(0, heights.max() * 1.2)

plt.tight_layout()

# ------------------------------------------
# 4) Animation
# ------------------------------------------
FRAMES = 40
PAUSE_FRAMES = 40   # so bleibt der finale Frame stehen â†’ "kein Loop"

def update(frame):
    if frame < FRAMES:
        scale = frame / (FRAMES - 1)
        for b, h in zip(bars, heights):
            b.set_width(h * scale)
    else:
        # nach Ende einfach finalen Zustand halten
        for b, h in zip(bars, heights):
            b.set_width(h)
    return bars

anim = FuncAnimation(
    fig, update,
    frames=FRAMES + PAUSE_FRAMES,
    interval=60, blit=False
)

# ------------------------------------------
# 5) Export GIF (quasi ohne Loop)
# ------------------------------------------
writer = PillowWriter(fps=30)
OUTPUT = "feature_importance_animation.gif"
anim.save(OUTPUT, writer=writer)

plt.close(fig)
print("Fertig â†’", OUTPUT)
```

```python
#!/usr/bin/env python3
"""
Interaktive SHAP-Landschaft (mit Z-Score Normalisierung der Feature-Werte)

- X-Achse: SHAP-Wert
- Y-Achse: Feature
- Farbe: z-Score der ursprÃ¼nglichen Feature-Werte
- Hintergrund: vollstÃ¤ndig transparent
"""

import shap
import pandas as pd
import numpy as np
import xgboost as xgb
import plotly.express as px
from pathlib import Path
from bootstrap import init as bootstrap_init

# ---------------------------------------------------
# 1) Config & Pfade
# ---------------------------------------------------
cfg = bootstrap_init(verbose=False)

tkey = cfg["defaults"]["target_species"]
ckey = cfg["defaults"]["contrast_species"]

t_pretty = cfg["species"][tkey]["name"].replace(" ", "_")
c_pretty = cfg["species"][ckey]["name"].replace(" ", "_")

model_path = Path(cfg["paths"]["output_dir"]) / tkey / f"model_MONTHLY_{t_pretty}_vs_{c_pretty}.json"
feature_csv = Path(cfg["paths"]["features_dir"]) / t_pretty / f"inat_with_climatology_{t_pretty}_vs_{c_pretty}.csv"

print("Model:", model_path)
print("CSV  :", feature_csv)

# ---------------------------------------------------
# 2) Modell und Daten laden
# ---------------------------------------------------
model = xgb.XGBClassifier()
model.load_model(str(model_path))

df = pd.read_csv(feature_csv)

valid_stats = ["ndvi_mean","ndwi_mean","moran_ndvi","geary_ndvi","moran_ndwi","geary_ndwi"]
feature_cols = [c for c in df.columns if c.startswith("m") and any(s in c for s in valid_stats)]

# Sampling fÃ¼r Performance
n = min(4000, len(df))
X = df[feature_cols].sample(n, random_state=0).reset_index(drop=True)

# ---------------------------------------------------
# 3) SHAP Werte berechnen
# ---------------------------------------------------
explainer = shap.TreeExplainer(model)
shap_vals = explainer.shap_values(X)

# In DataFrame bringen
rows = []
for i, feat in enumerate(feature_cols):
    vals = X[feat].values
    shapv = shap_vals[:, i]

    # Z-Score Normalisierung
    v_mean = np.nanmean(vals)
    v_std  = np.nanstd(vals)
    if v_std < 1e-6:
        zvals = np.zeros_like(vals)
    else:
        zvals = (vals - v_mean) / v_std

    for v, z, s in zip(vals, zvals, shapv):
        rows.append({
            "feature": feat,
            "value": v,
            "value_z": z,
            "shap": s
        })

df_shap = pd.DataFrame(rows)

# ---------------------------------------------------
# 4) Interaktiver Plotly-Beeswarm mit z-Score Farbskala
# ---------------------------------------------------
fig = px.scatter(
    df_shap,
    x="shap",
    y="feature",
    color="value_z",                 # z-Score statt raw value
    color_continuous_scale="Turbo",  # bessere Farbdynamik
    opacity=0.5,
    render_mode="webgl",
)

fig.update_traces(marker=dict(size=4))

fig.update_layout(
    title="Interaktive SHAP-Landschaft (z-Score normalisiert)",
    xaxis_title="SHAP value",
    yaxis_title="Feature",
    template="plotly_white",
    paper_bgcolor="rgba(0,0,0,0)",
    plot_bgcolor="rgba(0,0,0,0)",
)

# ---------------------------------------------------
# 5) HTML speichern
# ---------------------------------------------------
output_file = "shap_landscape_interactive_zscore.html"
fig.write_html(output_file, include_plotlyjs="cdn", full_html=True)

print("HTML gespeichert:", output_file)

fig.show()
```

```python
import json
import pandas as pd
from pathlib import Path

# Pfad zur .DUMP.txt-Datei
txt_dump = Path("/Volumes/Data/iNaturalist/outputs/macrolepiota_procera/model_MONTHLY_Macrolepiota_procera_vs_Parus_major.DUMP.txt")

# Alle Zeilen laden
lines = txt_dump.read_text().split("\n")

# --------------------------------------------------------
# 1) Splits extrahieren (feature + threshold)
# --------------------------------------------------------
import re

split_re = re.compile(
    r"\[(?P<feature>[a-z0-9_]+)<(?P<threshold>-?[0-9.]+)\]"
)

splits = []

for line in lines:
    m = split_re.search(line)
    if m:
        splits.append({
            "feature": m.group("feature"),
            "threshold": float(m.group("threshold"))
        })

df_splits = pd.DataFrame(splits)

print("Total splits:", len(df_splits))
df_splits.head()
```

```python
import re
import pandas as pd

def aggregate_seasonal_features(df_splits):
    """
    Gruppiert Monats-Features in saisonale Aggregationen:
    - Sommer: m07, m08, m09
    - Herbst: m10, m11, m12
    """
    
    def parse_feature(f):
        m = re.match(r"m(\d{2})_(.*)", f)
        if not m:
            return None, None
        month = int(m.group(1))
        base = m.group(2)
        
        if 7 <= month <= 9:
            season = "sommer"
        elif 10 <= month <= 12:
            season = "herbst"
        else:
            return None, None
        
        return season, base
    
    aggregated = {}
    
    for f in df_splits["feature"].unique():
        season, base = parse_feature(f)
        if season is None:
            continue
        key = f"{base}_{season}"
        aggregated.setdefault(key, []).append(f)
    
    return aggregated

aggregated_features = aggregate_seasonal_features(df_splits)
aggregated_features
```

```python
import numpy as np

# df_splits enthÃ¤lt alle Splits als (feature, threshold)

def aggregate_rules(df_splits, aggregated_features):
    """
    Bestimmt pro saisonal aggregiertem Featuretyp:
      - wie oft der Typ gesplittet wurde
      - alle Thresholds
      - reprÃ¤sentative Schwelle (median)
    """

    results = []

    for agg_name, raw_features in aggregated_features.items():
        # Splits filtern, die zu diesem Featuretyp gehÃ¶ren
        mask = df_splits["feature"].isin(raw_features)
        df_sub = df_splits[mask]

        if df_sub.empty:
            continue

        thresholds = df_sub["threshold"].values
        count = len(thresholds)
        median_thr = float(np.median(thresholds))

        results.append({
            "feature_type": agg_name,
            "count": count,
            "median_threshold": median_thr,
            "min_threshold": float(np.min(thresholds)),
            "max_threshold": float(np.max(thresholds)),
        })

    return pd.DataFrame(results).sort_values("count", ascending=False)


df_rules = aggregate_rules(df_splits, aggregated_features)
df_rules
```
